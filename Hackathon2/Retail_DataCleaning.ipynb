```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Ensure InvoiceDate is datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Snapshot date is the next day after the latest transaction
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# Compute Recency (days since last purchase per customer)
recency_df = df.groupby('CustomerID').agg({'InvoiceDate': lambda x: (snapshot_date - x.max()).days}).rename(columns={'InvoiceDate': 'Recency'})

# Compute Frequency (number of transactions per customer)
frequency_df = df.groupby('CustomerID').agg({'InvoiceNo': 'nunique'}).rename(columns={'InvoiceNo': 'Frequency'})

# Compute Monetary Value (total spending per customer)
monetary_df = df.groupby('CustomerID').agg({'TotalSales': 'sum'}).rename(columns={'TotalSales': 'Monetary'})

# Merge all RFM metrics
rfm_df = recency_df.merge(frequency_df, on='CustomerID').merge(monetary_df, on='CustomerID')

# Handle missing values if any
rfm_df.fillna(0, inplace=True)  # Replace NaN with 0

# KMeans Clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
rfm_df['Segment'] = kmeans.fit_predict(rfm_df[['Recency', 'Frequency', 'Monetary']])

# Plot Customer Segments
plt.figure(figsize=(8, 6))
plt.scatter(rfm_df['Frequency'], rfm_df['Monetary'], c=rfm_df['Segment'], cmap='viridis', alpha=0.6)
plt.xlabel('Frequency')
plt.ylabel('Monetary Value')
plt.title('Customer Segments based on Frequency & Monetary')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Drop rows with missing CustomerID and ensure it's an integer
df = df.dropna(subset=['CustomerID'])
df['CustomerID'] = df['CustomerID'].astype(int)

# Create user-item matrix (CustomerID vs StockCode with TotalSales as values)
user_item_matrix = df.pivot_table(index='CustomerID', columns='StockCode', values='TotalSales', fill_value=0)

# Compute cosine similarity between customers
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)

# Function to recommend items for a given user (with descriptions)
def recommend_items(customer_id, n=5):
    customer_id = int(customer_id)  # Ensure input is an integer

    # Check if customer exists
    if customer_id not in user_similarity_df.index:
        return f"Customer {customer_id} not found in the dataset."

    # Get similar users (excluding the user themselves)
    similar_users = user_similarity_df[customer_id].sort_values(ascending=False).index[1:]

    # Get products purchased by similar users
    recommended_items = df[df['CustomerID'].isin(similar_users)].groupby('StockCode')['TotalSales'].sum()

    # Get the top N recommended products
    top_n_products = recommended_items.nlargest(n)

    # Merge the descriptions for the recommended products
    product_descriptions = df[df['StockCode'].isin(top_n_products.index)][['StockCode', 'Description']].drop_duplicates()

    # Merge the total sales with descriptions
    recommendations_with_descriptions = pd.merge(top_n_products, product_descriptions, on='StockCode', how='left')

    return recommendations_with_descriptions[['Description', 'TotalSales']]

# Example: Recommend 5 products for Customer 17850
print(recommend_items(17850))
```
